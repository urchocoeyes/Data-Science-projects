```python
import pandas as pd
import numpy as np
```

```python
population = pd.read_csv('data/state-population.csv')
areas = pd.read_csv('data/state-areas.csv')
abbrevs = pd.read_csv('data/state-abbrevs.csv')
print(population.head()); print(areas.head()); print(abbrevs.head())
```

Нужно отсортировать штаты по плотности населения в 2010 году

```python
merged = pd.merge(population, abbrevs, how='outer', left_on='state/region', right_on='abbreviation').drop('abbreviation', axis=1
                                                                                                        )
merged
```

```python
merged.isnull()
```

```python
merged.isnull().any()
```

```python
merged[merged['state'].isnull()]
```

```python
merged.loc[merged['state'].isnull(), 'state/region']
```

```python
merged.loc[merged['state'].isnull(), 'state/region'].unique()
```

```python
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'
merged.isnull().any()
```

```python
final = pd.merge(merged, areas, on='state', how='left')
final
```

```python
final.isnull().any()
```

```python
final['state'][final['area (sq. mi)'].isnull()].unique()
```

```python
final.dropna(inplace=True)
final
```

```python
final.isnull().any()
```

# Snazzy

```python
data2010 = final.query("year == 2010 & ages == 'total'")
data2010
```

```python
data2010.set_index('state', inplace=True)
data2010
```

```python
density = data2010['population'] / data2010['area (sq. mi)']
density
```

```python
density.sort_values(ascending=False, inplace=True)
density.head()
```
